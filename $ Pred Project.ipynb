# **Scikit-learn Models**

```
from sklearn.utils.testing import all_estimators
from sklearn import base

import warnings
warnings.filterwarnings('ignore')

estimators = all_estimators()

for name, class_ in estimators:
    if issubclass(class_, base.RegressorMixin):
        print(name)
```

```
from sklearn.ensemble import RandomForestRegressor, ExtraTreesRegressor, BaggingRegressor, AdaBoostRegressor
from sklearn.ensemble import StackingRegressor, GradientBoostingRegressor, HistGradientBoostingRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.linear_model import LinearRegression, RidgeCV, SGDRegressor
from sklearn.svm import SVR
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import DotProduct, WhiteKernel 

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.metrics import mean_squared_error

from sklearn.pipeline import Pipeline
from joblib import dump
```

# **Using All Features**

```
X = df.drop('MEDV', axis=1)
y = df['MEDV']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```
names = ['Random Forest', 'Extra Trees', 'Bagging', 'AdaBoost', 'Stacking', 'GradientBoosting', 'HistGB', 'DecisionTree', 'KNeighborsR', 'MLPR', 
         'LinearRegression', 'RBF_SVR', 'Linear_SVR', 'GuassianProcessRegressor']

estimators = [('rf', RandomForestRegressor()), ('dt', DecisionTreeRegressor())]
kernel = DotProduct() + WhiteKernel()

models = [RandomForestRegressor(), ExtraTreesRegressor(), BaggingRegressor(), AdaBoostRegressor(), StackingRegressor(estimators=estimators),
          GradientBoostingRegressor(),HistGradientBoostingRegressor(), DecisionTreeRegressor(), KNeighborsRegressor(n_neighbors=3), 
          MLPRegressor(max_iter=750),LinearRegression(),SVR(),SVR(kernel='linear', gamma='auto'),GaussianProcessRegressor(kernel=kernel, alpha=1)]

acc = []
rmse = []

for i in models:
  model = i
  model.fit(X_train, y_train)
  y_pred = model.predict(X_test)
  acc.append(model.score(X_test, y_test)*100)
  rmse.append(np.sqrt(mean_squared_error(y_test, y_pred)))
```

```
d = {'Accuracy' : acc, 'RMSE': rmse}
progress = pd.DataFrame(d, index=names)
progress
```

```
plt.figure(figsize=(12,8))
progress['Accuracy'].plot(kind='barh', color='g')
plt.title('Accuracy & RMSE without Scaling Data', fontsize=25)
plt.yticks(fontsize=15)
plt.xticks(fontsize=15)
plt.figure(figsize=(12,8))
progress['RMSE'].plot(kind='barh', color='r')
plt.yticks(fontsize=15)
plt.xticks(fontsize=15)
plt.show()
```

# **Scaling Data**

```
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)
```

```
names = ['Random Forest', 'Extra Trees', 'Bagging', 'AdaBoost', 'Stacking', 'GradientBoosting', 'HistGB', 'DecisionTree', 'KNeighborsR', 'MLPR', 
         'LinearRegression', 'RBF_SVR', 'Linear_SVR', 'GuassianProcessRegressor', 'SGD']

estimators = [('rf', RandomForestRegressor()), ('dt', DecisionTreeRegressor())]
kernel = DotProduct() + WhiteKernel()

models = [RandomForestRegressor(), ExtraTreesRegressor(), BaggingRegressor(), AdaBoostRegressor(), StackingRegressor(estimators=estimators),
          GradientBoostingRegressor(),HistGradientBoostingRegressor(), DecisionTreeRegressor(), KNeighborsRegressor(n_neighbors=3), 
          MLPRegressor(max_iter=750),LinearRegression(),SVR(),SVR(kernel='linear', gamma='auto'),GaussianProcessRegressor(kernel=kernel, alpha=1),
          SGDRegressor()]

acc = []
rmse = []

for i in models:
  model = i
  model.fit(X_train, y_train)
  y_pred = model.predict(X_test)
  acc.append(model.score(X_test, y_test)*100)
  rmse.append(np.sqrt(mean_squared_error(y_test, y_pred)))
```

```
d = {'Accuracy' : acc, 'RMSE': rmse}
progress = pd.DataFrame(d, index=names)
progress
```

```
plt.figure(figsize=(12,8))
progress['Accuracy'].plot(kind='barh', color='g')
plt.title('Accuracy & RMSE with Scaled Data', fontsize=25)
plt.yticks(fontsize=15)
plt.xticks(fontsize=15)
plt.figure(figsize=(12,8))
progress['RMSE'].plot(kind='barh', color='r')
plt.yticks(fontsize=15)
plt.xticks(fontsize=15)
plt.show()
```

# **Choosing and Saving Model**

```
steps = [('scaling', StandardScaler()), ('Gradient Boosting Regressor', GradientBoostingRegressor(n_estimators=500, max_depth=6))]
model = Pipeline(steps)
model.fit(X_train, y_train)
print('Accuracy: {:.0f}%'.format(model.score(X_test, y_test) * 100))
```

```
dump(model, 'sklearn_model.pkl')
```

# **Keras Models**

```
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.optimizers import RMSprop
from sklearn.metrics import r2_score
```

```
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)
```

```
def model():
  model = Sequential()
  model.add(Dense(128, activation='relu', input_dim=13))
  model.add(Dropout(0.2))
  model.add(Dense(64, activation='relu'))
  model.add(Dropout(0.18))
  model.add(Dense(64, activation='relu'))
  model.add(Dense(1))
  
  model.compile(loss='mse', optimizer='adam')
  
  return model
```

```
model = model()
model.fit(X_train, y_train, shuffle=True, epochs=1000, verbose=0)
print('RMSE : {:.2f}'.format(np.sqrt(model.evaluate(X_test, y_test))))
```

```
y_pred = model.predict(X_test)
print(f'Accuracy: {r2_score(y_test, y_pred)*100:.2f}%')
```

```
model.save('keras_model.h5')
```
